```python
def convert(height):
    if 'cm' in height:
        return height
    else:
        feet, inches = height.split("\'")
        tot_inches = int(feet)*12 + int(inches.strip('"\n'))
        cm = tot_inches*2.54
        return f"{int(cm)}cm"
df['Height'] = df['Height'].apply(convert)
```

```python
import numpy as np
```

```python
''' Detection '''
# IQR
# Calculate the upper and lower limits
Q1 = df['OVA'].quantile(0.25)
Q3 = df['OVA'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
 
# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(df['OVA']>=upper)[0]
lower_array = np.where(df['OVA']<=lower)[0]
 
# Removing the outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
 
# Print the new shape of the DataFrame
print("New Shape: ", df.shape)
```
